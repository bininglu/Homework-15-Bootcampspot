```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pprint import pprint
from datetime import datetime
```

```python
tripdata_201801 = pd.read_csv("./Resources/201801-citibike-tripdata.csv")
tripdata_201802 = pd.read_csv("./Resources/201802-citibike-tripdata.csv")
tripdata_201803 = pd.read_csv("./Resources/201803-citibike-tripdata.csv")
tripdata_201804 = pd.read_csv("./Resources/201804-citibike-tripdata.csv")
tripdata_201805 = pd.read_csv("./Resources/201805-citibike-tripdata.csv")
tripdata_201806 = pd.read_csv("./Resources/201806-citibike-tripdata.csv")
tripdata_201807 = pd.read_csv("./Resources/201807-citibike-tripdata.csv")
tripdata_201808 = pd.read_csv("./Resources/201808-citibike-tripdata.csv")
tripdata_201809 = pd.read_csv("./Resources/201809-citibike-tripdata.csv")
tripdata_201810 = pd.read_csv("./Resources/201810-citibike-tripdata.csv")
tripdata_201811 = pd.read_csv("./Resources/201811-citibike-tripdata.csv")
tripdata_201812 = pd.read_csv("./Resources/201812-citibike-tripdata.csv")
```

```python
tripdata_201803.head()
```

```python
datalist = [tripdata_201801, tripdata_201802, tripdata_201803, tripdata_201804, tripdata_201805, tripdata_201806,
          tripdata_201807, tripdata_201808, tripdata_201809, tripdata_201810, tripdata_201811, tripdata_201812]
tripdata_monthly = pd.DataFrame()
for x in range(len(datalist)):
    tripdata_monthly.loc[x, 'month'] = str(x+1)
    tripdata_monthly.loc[x, 'total_trips'] = len(datalist[x])
    tripdata_monthly.loc[x, 'subscriber_trip_count'] = len(datalist[x].loc[datalist[x]["usertype"] == "Subscriber"])
    tripdata_monthly.loc[x, 'customer_trip_count'] = len(datalist[x].loc[datalist[x]["usertype"] == "Customer"])


tripdata_monthly.astype(int)
tripdata_monthly.to_csv("./Resources/Merge_tripdata_monthly.csv", index=False)
```

```python
bike_table = Merge_table_2018[['tripduration','bikeid']]
```

##Milage estimates are calculated using an assumed speed of 7.456 miles per hour, up to two hours. Trips over two hours max-out at 14.9 miles.

```python
bike_table.set_index("bikeid")

bike_table.groupby(["bikeid"])["tripduration"].sum()
# bike_table = bike_table.head(10)
bike_table["distance (miles)"] = bike_table["tripduration"]/3600*7.456

bike_table 
```

```python
Ave_distance=sum(bike_table["distance (miles)"])/len(bike_table["distance (miles)"])
```

```python
print(f'Average distance that a bike is {Ave_distance} miles')
```
